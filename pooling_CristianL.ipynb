## Importación de librerías

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Generación de matriz de escala de grises

```python
# Cargar la imagen en escala de grises
img = cv2.imread("number-6.png", cv2.IMREAD_GRAYSCALE)

# Aplicar umbral binario
_, img_binaria = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Convertir a matriz de 0s y 1s (0 = negro, 1 = blanco)
matriz = (img_binaria / 255).astype(np.uint8)
```

```python
# Mostrar la imagen con Matplotlib
plt.imshow(img, cmap='gray')
plt.axis('off')  # Ocultar ejes
plt.show()
```

```python
len(matriz)
```

## Pooling de 3x3 SUMA

```python
matriz2 = matriz.copy()
```

```python
length = int((len(matriz)/3)) # 20
```

```python
# Tamaño de la ventana de pooling
window_size = 3

# Calcular las dimensiones de la nueva matriz
new_rows = matriz2.shape[0] // window_size
new_cols = matriz2.shape[1] // window_size

# Crear la matriz de salida (20x20)
newMatrixP3 = np.zeros((new_rows, new_cols))

# Aplicar Max Pooling
for i in range(new_rows):
    for j in range(new_cols):
        # Extraer la ventana de 3x3
        window = matriz2[
            i * window_size : (i + 1) * window_size,
            j * window_size : (j + 1) * window_size
        ]
        newMatrixP3[i, j] = np.sum(window) # suma
        # newMatrixP3[i, j] = np.max(window) # máximo
        # newMatrixP3[i, j] = np.mean(window) # promedio

print("Matriz original:")
print(matriz2)
print("\nMatriz después de Suma Pooling:")
print(newMatrixP3)
```

```python
img2 = (newMatrixP3 * 255).astype(np.uint8)
# Guardar la imagen con OpenCV
cv2.imwrite("pool3_Suma.png", img2)
```

```python
import matplotlib.pyplot as plt
# Mostrar la imagen con Matplotlib
plt.imshow(img2, cmap='gray')
plt.axis('off')  # Ocultar ejes
plt.show()
```
