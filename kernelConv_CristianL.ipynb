```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = np.array([
     [10,20,30,40,50],
     [20,30,40,50,60],
     [30,40,50,60,70],
     [40,50,60,70,80],
     [50,60,70,80,90]], dtype=np.uint8)
```

```python
# Guardar la imagen con OpenCV
cv2.imwrite("MatrizOriginal.png", I)
```

```python
import matplotlib.pyplot as plt
# Mostrar la imagen con Matplotlib
plt.imshow(I, cmap='gray')
plt.axis('off')  # Ocultar ejes
plt.title("Matriz original")
plt.show()
```

```python
I2 = np.array([
      [80,80,80,0,0],
      [80,80,80,0,0],
      [80,80,80,0,0],
      [60,60,60,0,0],
      [20,20,20,0,0]], dtype=np.uint8)
```

```python
# Guardar la imagen con OpenCV
cv2.imwrite("MatrizKernel1.png", I2)
```

```python
import matplotlib.pyplot as plt
# Mostrar la imagen con Matplotlib
plt.imshow(I2, cmap='gray')
plt.axis('off')  # Ocultar ejes
plt.title("Matriz con primer kernel")
plt.show()
```
